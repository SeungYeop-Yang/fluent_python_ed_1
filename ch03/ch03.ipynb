# Dict Comprehensions

```python
DIAL_CODES = [
    (86, 'China'),
    (91, 'India'),
    (1, 'United States'),
    (62, 'Indonesia'),
    (55, 'Brazil'),
    (92, 'Pakistan'),
    (880, 'Bangladesh'),
    (234, 'Nigeria'),
    (7, 'Russia'),
    (81, 'Japan'),
    (82, 'Korea')
]

country_code = {country: code for code, country in DIAL_CODES}
country_code
```

```python
less_than_code_66 = {code: country.upper() for country, code in country_code.items() if code < 66}
less_than_code_66
```

```python
{code: less_than_code_66[code] for code in sorted(less_than_code_66)}
```

# Mapping with Flexble Key Lookup

Look at the attached python script: index_defaultdict.py

```python
class StrKeyDict0(dict):
    def __missing__(self, key):
        if isinstance(key, str):
            raise KeyError(key)
        return self[str(key)]
    
    def get(self, key, default=None):
        try:
            return self[key]
        except KeyError:
            return default
        
    def __contains__(self, key):
        return key in self.keys() or str(key) in self.keys()
```

```python
d = StrKeyDict0([('2', 'two'), ('4', 'four')])
d
```

```python
d['2']
```

```python
d['4']
```

```python
d[1]
```
